```python
import pandas as pd
df = pd.read_csv('data/data/driving_log.csv')
from IPython.display import display, HTML
display(df)
```

```python
num_bins = 200  # N of bins
bin_list = [None] * num_bins
start = 0
i = 0
for end in np.linspace(0, 1, num=num_bins):
    df_range = np.array(df[(np.absolute(df.steering) >= start) & (np.absolute(df.steering) < end)].index)
    bin_list[i] = df_range
    start = end
    i += 1
final_bin_list = bin_list
```

```python
batch_size = 32
nb_batches = 250

table_indices =[]
steer_values = []
for i in range(nb_batches):
    #print(i)
    j=0
    while j < batch_size:
        # get row index
        #start_index = np.random.randint(0,num_bins/2)
        start_index = 0
        row_index = np.random.randint(start_index, num_bins)
        num_vals_bin = len(final_bin_list[row_index])
        if num_vals_bin > 0:
            col_index = np.random.randint(num_vals_bin)
            table_index = final_bin_list[row_index][col_index]
            j += 1
            table_indices.append(table_index)
            #print(len(table_indices)) 
            steer = df.loc[table_index]['steering']
            if np.random.rand() < 0.5:
                steer *= -1
            steer_values.append(steer+(np.random.rand()-0.5)*0.1)

print(len(table_indices))
batch_samples = df.loc[table_indices]
```

```python
_=plt.hist(steer_values,200)
```
