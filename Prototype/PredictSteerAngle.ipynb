```python
def augment_brightness_camera_images(image):
    image1 = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    image1 = np.array(image1, dtype=np.float64)
    random_bright = 0.5 + np.random.uniform()
    image1[:, :, 2] = image1[:, :, 2] * random_bright
    image1[:, :, 2][image1[:, :, 2] > 255] = 255
    image1 = np.array(image1, dtype=np.uint8)
    image1 = cv2.cvtColor(image1, cv2.COLOR_HSV2RGB)
    return image1
```

```python
aug_img = augment_brightness_camera_images(crop_image_array)
plt.imshow(aug_img)
plt.savefig('./files/augumented_img.png')
```

```python
import cv2
def predict_val(file,model):
    image=cv2.imread(file)
    proc_image_array = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    final_image_array = proc_image_array[None,:,:,:]
    steering_angle = float(model.predict(final_image_array, batch_size=1))
    return steering_angle
```

```python
def predict_val2(file,model):
    image=cv2.imread(file)
    proc_image_array = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    crop_image_array = preprocess(proc_image_array,top_offset=60,bottom_offset=20)
    final_image_array = crop_image_array[None,:,:,:]
    steering_angle = float(model.predict(final_image_array, batch_size=1))
    return steering_angle
```
